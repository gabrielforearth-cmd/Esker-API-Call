```python
# Import the 'requests' library to make HTTP requests
import requests

# Define the URL of the API endpoint to obtain the token
url = "https://cc6.ondemand.esker.com/api/v1/token/apikey"

# Define the request headers, including the content type and authorization key
headers = {
    "Content-Type": "application/x-www-form-urlencoded",  # The content type of the request
    "Authorization": "*"  # The app authorization key
}

# Define the request payload containing the ApiKeyTag
data = {
    "ApiKeyTag": "*"  #  API key identifier for the request
}

# Make the POST request to the API with the provided URL, headers, and data
response = requests.post(url, headers=headers, data=data)

# Check if the response status code is 200 (OK)
if response.status_code == 200:
    # If successful, extract the 'access_token' from the response JSON
    token = response.json().get("access_token")
    print("Token:", token)  # Print the obtained token
else:
    # If the request failed, print the error status code and message
    print("Error:", response.status_code, response.text)
```
